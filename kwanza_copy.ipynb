## Customer Segmentation

```python
# perform segmentation of businesses based on the following criteria:
# Low-Value Customers: Total sales value less than 25th percentile
# Mid-Value Customers: Total sales value between 25th and 75th percentile
# High-Value Customers: Total sales value greater than 75th percentile
# Calculate the total sales value for each business
df_business_sales = df.groupby('ANONYMIZED BUSINESS').agg({'VALUE': 'sum'}).reset_index()

# Calculate the 25th and 75th percentiles of total sales value
value_25 = df_business_sales['VALUE'].quantile(0.25)
value_75 = df_business_sales['VALUE'].quantile(0.75)


# Perform segmentation based on the defined criteria
def customer_segmentation(value):
    if value < value_25:
        return 'Low-Value Customer'
    elif value < value_75:
        return 'Mid-Value Customer'
    else:
        return 'High-Value Customer'

df_business_sales['Customer Segment'] = df_business_sales['VALUE'].apply(customer_segmentation)
df_business_sales
```

```python
 #Forecast
df_forecast = model_fit.forecast(steps=len(test))

# Plot the results with specified colors
plt.figure(figsize=(14,7))
plt.plot(train.index, train["VALUE"], label='Train', color='#203147')
plt.plot(test.index, test["VALUE"], label='Test', color='#01ef63')
plt.plot(test.index, df_forecast, label='Forecast', color='orange')
plt.title('Value Forecast')
plt.xlabel('DATE')
plt.ylabel('Value')
plt.legend()
plt.show()
```

## 3. Anomaly Detection 

```python
#Anomaly Detection:Identify any unusual spikes or drops in sales performance (Quantity or Value) and explain possible reasons based on the data.
# Calculate the z-scores for Quantity and Value
df['Quantity_zscore'] = (df['QUANTITY'] - df['QUANTITY'].mean()) / df['QUANTITY'].std()
df['Value_zscore'] = (df['VALUE'] - df['VALUE'].mean()) / df['VALUE'].std()
# Identify unusual spikes or drops in sales performance (Quantity or Value)
anomalies_quantity = df[(df['Quantity_zscore'] > 3) | (df['Quantity_zscore'] < -3)]
anomalies_value = df[(df['Value_zscore'] > 3) | (df['Value_zscore'] < -3)]
# Print the anomalies
print('Anomalies in Quantity:')
print(anomalies_quantity)
print('Anomalies in Value:')
print(anomalies_value)
```

```python
#Create a dashboard summarizing key insights, including:
# Total Quantity and Value by Anonymized Category.
# Top-performing products and businesses.
# A time-series chart of sales trends.
# A segmentation summary of customer groups.

# Create a dashboard using Python libraries such as Plotly, Dash, or Streamlit to summarize key insights from the analysis. The dashboard should include visualizations of the following:
# Total Quantity and Value by Anonymized Category.
# Top-performing products and businesses.
# A time-series chart of sales trends.
# A segmentation summary of customer groups.
import dash
from dash import html
from dash import dcc
from dash.dependencies import Input, Output
import plotly.express as px
import pandas as pd
# Load the dataset provided
file_path = 'kwanza_refined.csv'
dataset = pd.read_csv(file_path)
df = pd.DataFrame(dataset)

# Create a dashboard using Dash
app = dash.Dash(__name__)


# Layout of the dashboard
app.layout = html.Div([
    html.H1("Kwanza Dataset Dashboard", style={'textAlign': 'center'}),

    # Dropdown to filter by a column (e.g., Product Category)
    dcc.Dropdown(
        id="category-dropdown",
        options=[{"label": cat, "VALUE": cat} for cat in df["ANONYMIZED CATEGORY"].unique()],
        placeholder="Select a Category",
        multi=True,
    ),

    # Line Chart for Sales Trends
    dcc.Graph(id="sales-trend"),

    # Bar Chart for Top Products
    dcc.Graph(id="top-products"),
])

# Callback to update charts based on dropdown selection
@app.callback(
    [Output("sales-trend", "figure"), Output("top-products", "figure")],
    [Input("category-dropdown", "VALUE")]
)
def update_charts(selected_categories):
    filtered_df = df.copy()

    # Apply filter if selection is made
    if selected_categories:
        filtered_df = filtered_df[filtered_df["ANONYMIZED CATEGORY"].isin(selected_categories)]

    # Sales Trend (Line Chart)
    fig1 = px.line(
        filtered_df,
        x="DATE",
        y="VALUE",
        title="Sales Trend Over Time",
        markers=True
    )

    # Top Products (Bar Chart)
    fig2 = px.bar(
        filtered_df.groupby("ANONYMIZED PRODUCT")["VALUE"].sum().nlargest(10).reset_index(),
        x="ANONYMIZED PRODUCT",
        y="VALUE",
        title="Top 10 Products by Revenue",
        text_auto=True
    )

    return fig1, fig2

# Run the app
if __name__ == "__main__":
    app.run_server(debug=True)

#Running the server slowing down my computer, so I will stop it here.
```
