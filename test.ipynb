```python
import pandas as pd
# Load the dataset provided 
file_path = 'kwanza.csv'
dataset = pd.read_csv(file_path)


# Display the first few rows of the dataset to understand its structure
dataset.head()
```

```python
df = pd.DataFrame(dataset)
```

```python
df.head()
```

```python
df.info()
```

```python
df["UNIT PRICE"] = df["UNIT PRICE"].astype(str).str.replace(",", "")
```

```python
df.head()
```

```python
df["UNIT PRICE"] = df["UNIT PRICE"].astype(float)

df.head()
```

```python
#convert quantity column from integer to float
df["QUANTITY"] = df["QUANTITY"].astype(float)

df.head()
```

```python
df = df.dropna()  # Remove rows with missing values
df = df.drop_duplicates()  # Remove duplicate rows
df.info()
```

```python
df['DATE'] = pd.to_datetime(df['DATE'])
df['Month-Year'] = df['DATE'].dt.strftime('%B %Y')
df.head()
```

```python
df_category = df.groupby('ANONYMIZED CATEGORY').agg({'QUANTITY': 'sum', 'UNIT PRICE': 'sum'}).reset_index()
df_category
# Calculate the total Quantity and Value grouped by Anonymized Business
df_business = df.groupby('ANONYMIZED BUSINESS').agg({'QUANTITY': 'sum', 'UNIT PRICE': 'sum'}).reset_index()
df_business
```

```python
df_category
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Visualize the total Quantity and Value grouped by Anonymized Category
#plt.figure(figsize=(12, 6))
plt.figure(figsize=(22, 22))
plt.subplot(1, 2, 1)
sns.barplot(x='QUANTITY', y='ANONYMIZED CATEGORY', data=df_category, )
plt.title('Total Quantity by Anonymized Category')
```

```python
# Caluculate the total Value grouped by Anonymized Business by multiplying Quantity and Unit Price
df_business['VALUE'] = df_business['QUANTITY'] * df_business['UNIT PRICE']
# Group by Anonymized Business and sum the total Value
df_business = df_business.groupby('ANONYMIZED BUSINESS').agg({'QUANTITY': 'sum', 'VALUE': 'sum'}).reset_index()
df_business
```
